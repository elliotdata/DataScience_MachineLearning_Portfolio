```python
#importing necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

#base_url and total_number of pages to be iterated

base_url = "https://www.airlinequality.com/airline-reviews/british-airways/"
total_pages = 363 #total 363 pages on the website

#Variable to store the output
reviews_list = []

#for loop to iterate through the pages
for page_num in range(1, total_pages + 1):
    page_url = f"{base_url}page/{page_num}/"
    
#if else statement to find
    response = requests.get(page_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        reviews = soup.find_all("div", class_="text_content") #obtained by inspecting the html code

        print(f"Scraping reviews from page {page_num}")

        for review in reviews:
            review_text = review.get_text(strip=True)
            # Remove unwanted text patterns
            review_text = review_text.replace("âœ…Trip Verified|", "").replace("Not Verified|", "")
            reviews_list.append(review_text)

    else:
        print(f"Failed to fetch page {page_num}")

print("Scraping complete.")

# Creating a DataFrame
data = {'review': reviews_list}
df = pd.DataFrame(data)
```

**STANDARDIZATION**

```python
#creating a function to standardize the rows in a column
column_to_standardize = ['review']

# Function to remove special characters, leading/trailing spaces, and convert to title case
def standardize_column(df, column_name):
   df[column_name] = df[column_name].str.replace('[^a-zA-Z0-9\s]', '', regex=True) #removes special characters
   df[column_name] = df[column_name].str.strip() #strips of whitespaces
   df[column_name] = df[column_name].str.lower() #lowercase


#Applying standardize function
for column in column_to_standardize:
  standardize_column(df, column)
```

```python
#saving the output file
df.to_csv('BA_reviews.csv',index=False)  
```
